```python

# Program to remove all whitespaces
import re

# multiline string
string = 'abc 12\
de 23 \n f45 6'

# matches all whitespace characters
pattern = '\s+'

# empty string
replace = ''

new_string = re.sub(pattern, replace, string) 
print(new_string)
```

```python
import re

string = "Python is fun"

# check if 'Python' is at the beginning
match = re.search('\APython', string)

if match:
  print("pattern found inside the string")
else:
  print("pattern not found")  
```

```python
import re

string = '39801 356 2102 1111'

# Three digit number followed by space followed by two digit number
pattern = '(\d{3}) (\d{2})'

# match variable contains a Match object.
match = re.search(pattern, string) 

if match:
  print(match.group())
else:
  print("pattern not found")

print(match.group(1))
print(match.group(2))
```

```python
import mysql.connector
conndb = mysql.connector.MySQLConnection( host="localhost",user="root",password="password",database="bnp")

sqlcursor=conndb.cursor(dictionary=True)
sqlcursor.execute("select * from customer")
result = sqlcursor.fetchall()
from lxml import etree
root=etree.Element("customers")
for res in result:
    cust = etree.Element("customer")
    for k,v in res.items():
        ele = etree.Element(k)
        ele.text = str(v)
        cust.append(ele)
    root.append(cust)
print(etree.tostring(root,pretty_print=True).decode("utf-8"))
```
