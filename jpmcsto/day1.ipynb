```python
from lxml import etree
root= etree.Element("email")
frome=etree.Element("from")
frome.text ="xyz@gmail.com"
to=etree.Element("to")
for i in range(1,10):
    toaddress=etree.SubElement(to,"toaddress")
    toaddress.text ='abc{}@gmail.com'.format(i)
    if i % 2 == 0:
        toaddress.set('type','bcc')
    else:
        toaddress.set('type','cc')
    
root.append(frome)
root.append(to)
print(etree.tostring(root,pretty_print=True).decode())
```
